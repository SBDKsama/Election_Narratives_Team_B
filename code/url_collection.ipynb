```python
import json
from urllib.parse import urlparse
from collections import OrderedDict


# Load the JSON data from the file
with open('../data/emails_extracted.json', 'r') as file:
    emails = json.load(file)

# with open('../data/hostname.json', 'r') as file:
#     hostnames = json.load(file)

# Initialize a dictionary for hostname counts
hostname_counts = {}

for email in emails: 
    for url in email["urls"]: 
        try: 
            # Parse the URL to extract the hostname
            parsed_url = urlparse(url)
            hostname = parsed_url.hostname

            # Increment the count for this hostname
            if hostname:
                hostname_counts[hostname] = hostname_counts.get(hostname, 0) + 1
        except ValueError: 
            print()

# Write the updated hostname counts to 'hostname.json'
with open('../data/hostname.json', 'w') as file:
    json.dump(hostname_counts, file, indent=4)
```
